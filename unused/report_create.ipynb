```python
import pandas as pd
import numpy as np
```

```python
requests = pd.read_csv(r"C:\Users\Vadim\PycharmProjects\ScrapMachine\data\requests\requests_2021-05_data.csv",
                       header=0, delimiter=';', encoding='cp1251')
requests.columns = ['date', 'company', 'client', 'message', 'reactions', 'decision', 'specialist', 'dec_in_time',
                    'react', 'react_in_time', 'react_avg']

requests.head(50)
```

```python
data_on_specs = requests.groupby(['specialist'], as_index=False, dropna=False)\
    .aggregate({'message': 'count',
                'dec_in_time': 'sum',
                'react': 'sum',
                'react_in_time': 'sum',
                'decision': 'mean',
                'react_avg': 'mean'})\
    .rename(columns= {'message': 'req_count', 'decision': 'dec_avg'})

data_on_specs['dec_in_time_perc'] = round(data_on_specs['dec_in_time'] / data_on_specs['req_count'] * 100, 2)
data_on_specs['react_in_time_perc'] = round(data_on_specs['react_in_time'] / data_on_specs['react'] * 100, 2)

data_on_specs = data_on_specs.rename(columns={'specialist': 'Специалист',
                                              'req_count': 'Запросов всего',
                                              'dec_in_time': 'Решено в срок',
                                              'react': 'Реакций всего',
                                              'react_in_time': 'Реакций в срок',
                                              'dec_avg': 'Время решения ср.',
                                              'react_avg': 'Время реакции ср.',
                                              'dec_in_time_perc': 'Время решения %',
                                              'react_in_time_perc': 'Скорость реакции %'})

data_on_specs.head()
```

```python
data_on_specs.to_excel(r"C:\Users\Vadim\PycharmProjects\ScrapMachine\data\requests\requests_2021-05_report.xlsx",
                       sheet_name='2021-05_report', index = False)
```
